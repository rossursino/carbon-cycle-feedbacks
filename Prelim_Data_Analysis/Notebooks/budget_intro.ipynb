```python
import numpy as np
from scipy import stats as sp
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
df = pd.read_csv('./../gcb_data/budget.csv')
df.head()
```

```python
year = df['Year']
ffi = df['fossil fuel and industry']
lce = df['land-use change emissions']
atmg = df['atmospheric growth']
os = df['ocean sink']
ls = df['land sink']
imb = df['budget imbalance']
```

```python
plt.plot(year, -ls)
axes = plt.gca()
axes.set_ylim([-7,2])
plt.title('GCB Land Flux')
plt.xlabel('Year')
plt.ylabel('C Flux To The Atmosphere (GtC/yr)')
plt.savefig('./../../Peylin_flux/Output/GCB_land.png')
```

```python
plt.plot(year, -os)
axes = plt.gca()
axes.set_ylim([-6,3])
plt.title('GCB Ocean Flux')
plt.xlabel('Year')
plt.ylabel('C Flux To The Atmosphere (GtC/yr)')
plt.savefig('./../../Peylin_flux/Output/GCB_ocean.png')
```

```python
# Plot of ocean and land sink in the last 60 years.
plt.plot(year, -os)
plt.plot(year, -ls)
plt.legend()
```

```python
plt.plot(year, ffi)
plt.plot(year, lce)
axes = plt.gca()
axes.set_ylim([0,10])
```

```python
# Plot of airborne fraction (growth/emissions) in the last 60 years.
airborne = atmg/(ffi+lce)
plt.plot(year, airborne)
```

```python
plt.plot(year, atmg)
```

```python
plt.plot(year, imb)
```

```python
plt.plot(year,ffi+lce)
plt.plot(year,atmg)
plt.legend(['emissions', 'atm-growth'])
```

```python
plt.plot(year,os+ls)
plt.plot(year,atmg)
plt.legend(['total_uptake', 'atm-growth'])
```
