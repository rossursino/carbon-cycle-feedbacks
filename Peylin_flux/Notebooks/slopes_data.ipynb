```python
import pickle
import pandas as pd
```

```python
Rayner_reg_land = pickle.load(open('./../Output/output_linreg/Rayner/regression_land.pik', 'rb'))
Rayner_reg_land
```

```python
RRL = pd.DataFrame(Rayner_reg_land).iloc[0]; RRL
```

```python
def slopes(fname):
    data = pickle.load(open('./../Output/output_linreg/'+fname, 'rb'))
    return pd.DataFrame(data).iloc[0]
```

```python
RRO = slopes('Rayner/regression_ocean.pik'); RRO
```

```python
pd.DataFrame({'Rayner_land': RRL, 'Rayner_ocean': RRO})
```

```python
CAMS_RL = slopes('CAMS/regression_land.pik'); CAMS_RL
```

```python
JAMSTEC_RL = slopes('JAMSTEC/regression_land.pik'); JAMSTEC_RL
```

```python
pd.DataFrame({'Rayner_land': RRL, 'Rayner_ocean': RRO, 'CAMS_land': CAMS_RL, 'JAMSTEC_land':JAMSTEC_RL})
```

```python
land_fnames = ['Rayner/regression_land.pik', 'CAMS/regression_land.pik', 'JAMSTEC/regression_land.pik',
               'JENA_s76/regression_land.pik', 'JENA_s85/regression_land.pik', 'CTRACKER/regression_land.pik',
               'GCP/regression_land.pik']
ocean_fnames = ['Rayner/regression_ocean.pik', 'CAMS/regression_ocean.pik', 'JAMSTEC/regression_ocean.pik',
                'JENA_s76/regression_ocean.pik', 'JENA_s85/regression_ocean.pik', 'CTRACKER/regression_ocean.pik',
                'GCP/regression_ocean.pik']
```

```python
land_fnames[4][:-20]
```

```python
land_slopes = {}
for fname in land_fnames:
    land_slopes[fname[:-20]] = slopes(fname)

land_df = pd.DataFrame(land_slopes)
land_df
```

```python
ocean_slopes = {}
for fname in ocean_fnames:
    ocean_slopes[fname[:-21]] = slopes(fname)

ocean_df = pd.DataFrame(ocean_slopes)
ocean_df
```

```python
land_df.to_csv('./../Output/output_linreg/reg_land_slopes.csv')
ocean_df.to_csv('./../Output/output_linreg/reg_ocean_slopes.csv')
```
