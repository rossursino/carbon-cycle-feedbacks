```python
import matplotlib.pyplot as plt
import pandas as pd
import sys
import scipy.stats as stats
```

```python
fGCB = './../../Prelim_Data_Analysis/gcb_data/budget.csv'
df = pd.read_csv(fGCB,
                 index_col=0,
                 usecols=[0,4,5])
df.head()
```

```python
df.plot()
```

```python
ocean = -df["ocean sink"]
land = -df["land sink"]
```

```python
plt.plot(ocean)
```

```python
def lin_reg(time_list, data):
    
    regression = {}
    for period in time_list.keys():
        x=time_list[period]
        y=data[x].values

        reg = stats.linregress(x,y)
        slope=reg[0]
        intercept=reg[1]
        
        line = slope*x+intercept
        
        regression[period] = reg

        plt.plot(x, line)
    
    return regression
```

```python
x1=ocean.index[1:12]
x2=ocean.index[11:22]
x3=ocean.index[21:32]
x4=ocean.index[31:42]
x5=ocean.index[41:52]
x6=ocean.index[48:59]
```

```python
x4
```

```python
time_dict={'1960-1970':x1, '1970-1980':x2, '1980-1990':x3, '1990-2000':x4, '2000-2010':x5, '2007-2017':x6}
```

```python
plt.figure(figsize=(10,6))
plt.plot(ocean)
lin_reg(time_dict, ocean)
plt.title('GCB Ocean Uptake with decadal trends')
plt.xlabel('Year')
plt.ylabel('C flux to the atmosphere (GtC/yr)')
```

```python
plt.figure(figsize=(10,6))
plt.plot(land)
lin_reg(time_dict, land)
plt.title('GCB Land Uptake with decadal trends')
plt.xlabel('Year')
plt.ylabel('C flux to the atmosphere (GtC/yr)')
```

```python
plt.plot(land.index, land)
plt.plot(ocean.index, ocean)
plt.title('GCB Uptake: 1959-2017')
plt.xlabel('Year')
plt.ylabel('C flux to the atmosphere (GtC/yr)')
```
