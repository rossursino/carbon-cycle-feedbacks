```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
df = pd.read_csv('./../gcb_data/historical_budget.csv')
df.head()
```

```python
year = df['Year']
ffi = df['fossil fuel and industry']
lce = df['land-use change emissions']
atmg = df['atmospheric growth']
os = df['ocean sink']
ls = df['land sink']
imb = df['budget imbalance']
```

Plots of ocean and land sink in the Industrial Era.

```python
plt.plot(year, os)
```

```python
plt.plot(year, ls)
```

```python
plt.plot(year, os)
plt.plot(year, ls)
plt.legend()
```

```python
# Plot of airborne fraction (growth/emissions) in the Industrial Era.
airborne = atmg/(ffi+lce)
plt.plot(year, airborne)
```

```python
plt.plot(year, atmg)
```

```python
plt.plot(year, imb)
```

```python
plt.plot(year,ffi+lce)
plt.plot(year,atmg)
plt.legend(['emissions', 'atm-growth'])
```
