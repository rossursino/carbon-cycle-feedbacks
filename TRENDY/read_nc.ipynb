Notebook to access nc files and plot values using cartopy

```python
import xarray as xr
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import numpy as np
```

ENTER NAME OF FILE BELOW:

```python
fname = 'LPJ-GUESS/S3/LPJ-GUESS_S3_cVeg.nc'
```

----

OPEN FILE USING XARRAY

```python
file = xr.open_dataset(fname)
```

```python
file
```

----

FUNCTIONS

```python
def single_time_plot(time):
    Veg = file.cVeg.sel(time=time).values[0]
    
    fig = plt.figure(figsize=(15,10))
    ax = fig.add_subplot(1,1,1,projection=ccrs.PlateCarree())
    ax.coastlines()
    ax.contourf(file.longitude, file.latitude, Veg)
```

```python
def single_point_timeseries(lat, lon):
    Veg = file.cVeg.sel(latitude=lat, longitude=lon)
    
    plt.plot(np.arange(1700, 2019), Veg.values)
```

----

IMPLEMENTATION

```python
single_time_plot("2018")
```

```python
single_point_timeseries(lat=9.25, lon=9.25)
```
